```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# from sklearn.datasets.samples_generator import make_blobs
from sklearn.datasets import make_blobs

x,y_label=make_blobs(n_samples=300 , centers=3 , cluster_std = 0.9 , random_state= 0)
plt.scatter(x=x[:,0] , y=x[:,1] , s=30 );
```

```python
from sklearn.cluster import KMeans
kmean = KMeans(init='random' , n_clusters=3 , random_state= 42)
kmean.fit(x)
y_kmean = kmean.predict(x)

plt.scatter(x=x[:,0]  , y=x[:,1] , c=y_kmean , s=30 );
center = kmean.cluster_centers_
plt.scatter(x=center[:,0]  , y=center[:,1] , c='red' , s=100 , alpha=0.5 );
```

```python
# choosing approve number of clusters

kkmeans_kwargs = {'init': 'random' , 'random_state': 42 }
count = []
number_test_k = 20
for k in range(1,number_test_k) :
    kkmean= KMeans( n_clusters=k , **kkmeans_kwargs )
    kkmean.fit(x)
    count.append(kkmean. inertia_)
# print(count)

plt.plot(range(1,number_test_k) , count)
plt.xticks(range(1,number_test_k))
plt.xlabel('Number of cluster')
plt.ylabel('Loss function')
plt.show()
```
