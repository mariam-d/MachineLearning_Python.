```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

%matplotlib inline
```

```python
# import inline dataset
from sklearn.datasets import load_breast_cancer
data = load_breast_cancer()   # data is dictionary
```

```python
# data.keys()           # dict_keys(['data', 'target', 'target_names', 'DESCR', 'feature_names', 'filename'])
# data['data']
# data['target']
# data['feature_names']
```

```python
df = pd.DataFrame( data = data['data'] , columns = data['feature_names'])
```

```python
# preprocessing => scale data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)
```

```python
# PCA
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
new_data = pca.fit_transform(scaled_data)

# scaled_data.shape     # (569, 30)
# new_data.shape        # (569, 2)
```

```python
# lets plot two dimensions
plt.figure(figsize=(8,6))
plt.scatter(new_data[:,0] , new_data[:,1] , c=data['target'] , cmap='rainbow')

plt.xlabel('first principal component')
plt.ylabel('second principal component')
```

```python
#plot explained variances
plt.figure(figsize=(8,6))
plt.bar( range(pca.n_components_) , pca.explained_variance_)

plt.xticks(range(pca.n_components_))
plt.xlabel('PCA features')
plt.ylabel('Variance')
plt.show()
```
