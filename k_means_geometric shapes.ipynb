```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# from sklearn.datasets import make_moons
from sklearn.datasets.samples_generator import make_moons
x , y = make_moons(n_samples=1300 , noise=0.05 , random_state= 0)
plt.scatter(x = x[:,0] , y = x[:,1] , s = 50 );
```

```python
# Kmeans => Linear 

from sklearn.cluster import KMeans
labels = KMeans(init='random' , n_clusters=2 , random_state= 42).fit_predict(x)
plt.scatter(x=x[:,0]  , y=x[:,1] , c=labels , s=30 );
```

```python
# Kmeans => geometric shapes

from sklearn.cluster import SpectralClustering
labels = SpectralClustering( n_clusters=2 , affinity = 'nearest_neighbors' , assign_labels='kmeans').fit_predict(x)
plt.scatter(x=x[:,0]  , y=x[:,1] , c=labels , s=30  );
```
